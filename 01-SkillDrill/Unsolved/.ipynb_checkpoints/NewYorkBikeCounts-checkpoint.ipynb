```python
# Dependencies
#%pip install statsmodels
import pandas as pd
import plotly.express as plt
import os
import statsmodels.api as sm
import numpy as np
```

```python
bikecounts_file = os.path.join('../','Resources', 'bikecounts.csv')
bikecounts_df = pd.read_csv(bikecounts_file)
bikecounts_df['Day'] = pd.to_datetime(bikecounts_df['Day'])
bikecounts_df.head(30)
```

```python
bikecounts_df['Brooklyn Bridge'].sum()
```

```python
bikecounts_df.info()
```

```python
#pd.to_numeric(bikecounts_df['Precipitation'],downcast='float')
#bikecounts_df['Precipitation'] = bikecounts_df['Precipitation'].astype(float)
bikecounts_df['Precipitation'] = pd.to_numeric(bikecounts_df['Precipitation'], errors='coerce')
bikecounts_df = bikecounts_df.replace(np.nan, 0, regex=True)
bikecounts_df.info()
```

```python
bikecounts_df.head()
```

```python
X = bikecounts_df[["High Temp (°F)","Low Temp (°F)","Precipitation"]]
y = bikecounts_df['Brooklyn Bridge']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
```

```python
print(results.summary())
```

**My Analysis:
    
> Null Hypothesis: The number of bikes crossing Brooklyn Bridge depends on both the Temperature(Low&High) and the Precipitation.

> Alternative Hypothesis: The number of bikes crossing Brooklyn Bridge is independent of both the Temperature(Low&High) and the Precipitation.

**Understanding the OLS Regression Results:

We can say that, the correlation between any two columns in the dataframe is statistically significant when the p-value < 0.05. As the probability that the null is correct is less than 5%, it is strongly against the null hypothesis meaning that the results are randomly generated. 

> Precipitation: Pvalue = 0.012 denotes that there is a statistically significant correlation between the bikes crossing the Brooklyn Bridge and the precipitation, as the p-value < 0.05.

> Low Temp (°F): Pvalue = 0.299 Here the p-value > 0.05 and hence is not as significant in determining the number of bikes crossing the bridge.

> High Temp (°F): Pvalue = 0.000 In this case, the p-value < 0.05 which denotes that this parameter has a statistically significant correlation with the bikes crossing the Brooklyn bridge.

It is due to this analysis, that we consider the alternative hypothesis and neglect the null hypothesis.
